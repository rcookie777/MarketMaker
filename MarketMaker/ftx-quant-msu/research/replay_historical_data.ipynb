```python
columns = {'bid':{'price': bid_price_columns,
                 'amount': bid_amount_columns},
           'ask':{'price': ask_price_columns,
                  'amount': ask_amount_columns}}
```

```python
def get_price_and_amounts(index, side, df, columns):
    price_cols, amount_cols = get_price_and_amounts(side, columns)
    prices = df.loc[index, price_cols]
    amounts = df.loc[index, amount_cols]
    return prices, amounts
    
def get_price_and_amount_columns(side, columns):
    return columns[side]['price'], columns[side]['amounts']
```

```python
def animate(ii, df, ax, columns):
    '''
    ask_prices, ask_amounts = get_price_and_amounts(ii, 'ask', df, columns)
    bid_prices, bid_amounts = get_price_and_amounts(ii, 'bid', df, columns)
    
    ask_prices, ask_amounts = get_price_and_amounts('ask', columns)
    bid_prices, bid_amounts = get_price_and_amounts('bid', columns)
    '''

    bid_prices = df.loc[ii, bid_price_columns]
    bid_amounts = df.loc[ii, bid_amount_columns]
    ask_prices = df.loc[ii, ask_price_columns]
    ask_amounts = df.loc[ii, ask_amount_columns]
    
    ask_snap_df = pd.DataFrame(columns=["prices", "amounts"])
    ask_snap_df["prices"] = ask_prices.to_numpy()
    ask_snap_df["amounts"] = ask_amounts.to_numpy()

    bid_snap_df = pd.DataFrame(columns=["prices", "amounts"])
    bid_snap_df["prices"] = bid_prices.to_numpy()
    bid_snap_df["amounts"] = bid_amounts.to_numpy()
    
    ax.clear()
    sns.ecdfplot(x="prices",
                weights="amounts",
                stat="count",
                 data=bid_snap_df,
                 complementary=True,
                 color="green",
                 ax=ax
                )

    sns.ecdfplot(x="prices",
                weights="amounts",
                stat="count",
                 data=ask_snap_df,
                 color="red",
                 ax=ax
                )
```
